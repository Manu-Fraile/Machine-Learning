```python
import numpy as np
import pandas as pd
import math as m
import matplotlib.pyplot as plt
import seaborn as sns
import csv

%matplotlib inline
```

```python
evaluationDF = pd.read_csv('EvaluateOnMe.csv')
trainingDF = pd.read_csv('TrainOnMe.csv')
```

```python
w_data = []
col = ''

#Cleaning wrong values
for i in range(len(trainingDF)):
        
    for j in range(len(trainingDF.iloc[i])):
        
        if (j!=1) and (j!=6) and (j!=7):
            
            datum = trainingDF.iloc[i][j]
            
            if type(datum) is not np.float64:
                
                if j==0: col = 'id'                   
                else: col = 'x'+str(j-1)
                
                try:
                    datum = np.float64(datum)
                except:
                    w_data.append(i)
                    
                trainingDF[col][i] = datum
                
        elif j==6:
            if trainingDF['x5'][i]=='True':
                trainingDF['x5'][i]=1.0
                
            elif trainingDF['x5'][i]=='False':
                trainingDF['x5'][i]=0.0
                
            else:
                w_data.append(i)
    
w_data = list(dict.fromkeys(w_data))
trainingDF.drop(w_data,inplace=True)
trainingDF = trainingDF.reset_index(drop=True)

#Reset data types
for k in range(11):
    if (k!=1) and (k!=7):
        if k==0: 
            col = 'id'  
            trainingDF[col] = trainingDF[col].astype('int64')
        else: 
            col = 'x'+str(k-1)
            trainingDF[col] = trainingDF[col].astype('float64')
            
#trainingDF['x5'] = trainingDF['x5'].astype('bool')

non_data=[]

#Clean non-representative data
for i in range(len(trainingDF)):
    
    for j in range(len(trainingDF.iloc[i])):
        if (j!=1) and (j!=6) and (j!=7):
            
            if j==0: col = 'id'                   
            else: col = 'x'+str(j-1)
            
            if m.isnan(trainingDF[col][i]):
                non_data.append(i)
                    
    if trainingDF['x7'][i]<-200:
        non_data.append(i)
        
    if trainingDF['x8'][i]>50000:
        non_data.append(i)
        

non_data = list(dict.fromkeys(non_data))
trainingDF.drop(non_data,inplace=True)
trainingDF = trainingDF.reset_index(drop=True)


#Converting T-F to int & grades to int
for i in range(len(trainingDF)):
    
    #if trainingDF['x5'][i]:
    #    trainingDF['x5'][i] = 1.0
    #else:
    #    trainingDF['x5'][i] = 0.0

        
    if trainingDF['x6'][i]=='A': trainingDF['x6'][i]=5.0
    if trainingDF['x6'][i]=='B': trainingDF['x6'][i]=4.0
    if trainingDF['x6'][i]=='C': trainingDF['x6'][i]=3.0
    if trainingDF['x6'][i]=='D': trainingDF['x6'][i]=2.0
    if trainingDF['x6'][i]=='E': trainingDF['x6'][i]=1.0
    if trainingDF['x6'][i]=='Fx': trainingDF['x6'][i]=0.5
    if trainingDF['x6'][i]=='F': trainingDF['x6'][i]=0.0
        
trainingDF['x6'] = trainingDF['x6'].astype('float64')
```

# the model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
accuracy = 0

for i in range(800):
    
    X = trainingDF.drop(['id', 'y'],axis=1)
    y = trainingDF['y']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
    
    rfc = RandomForestClassifier(n_estimators=800)
    
    rfc.fit(X_train,y_train)
    
    predictions = rfc.predict(X_test)
    
    if accuracy_score(y_test,predictions) > accuracy:
        
        our_model = rfc
        
        accuracy = accuracy_score(y_test,predictions)
```

```python
predictions = our_model.predict(X_test)
print(classification_report(y_test,predictions))
```

# apply the model

```python
#Converting T-F to int & grades to float
for i in range(len(evaluationDF)):
    
    if evaluationDF['x5'][i]==True:
        evaluationDF['x5'][i]=1.0
                
    elif evaluationDF['x5'][i]==False:
        evaluationDF['x5'][i]=0.0

        
    if evaluationDF['x6'][i]=='A': evaluationDF['x6'][i]=5.0
    if evaluationDF['x6'][i]=='B': evaluationDF['x6'][i]=4.0
    if evaluationDF['x6'][i]=='C': evaluationDF['x6'][i]=3.0
    if evaluationDF['x6'][i]=='D': evaluationDF['x6'][i]=2.0
    if evaluationDF['x6'][i]=='E': evaluationDF['x6'][i]=1.0
    if evaluationDF['x6'][i]=='Fx': evaluationDF['x6'][i]=0.5
    if evaluationDF['x6'][i]=='F': evaluationDF['x6'][i]=0.0
        
evaluationDF['x6'] = evaluationDF['x6'].astype('float64')
evaluationDF['x5'] = evaluationDF['x5'].astype('float64')
```
